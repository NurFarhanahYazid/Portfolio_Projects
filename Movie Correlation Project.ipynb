```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#read data

df= pd.read_csv(r'C:\Users\nurfa\Documents\Data Analyst Bootcamp\Python\movies.csv')
```

```python
df.head()
```

```python
#check missing data

for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print ('{} - {}%'.format(col,pct_missing))
```

```python
df = df.dropna()
```

```python
for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print ('{} - {}%'.format(col,pct_missing))
```

```python
#data type - columns

df.dtypes
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int)
```

```python
df = df.astype({col: 'int64' for col in df.select_dtypes('int32').columns})
```

```python
df.dtypes
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
#display all the data

pd.set_option('display.max.rows', None)
```

```python
#drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
#scatter plot with budget vs gross

plt.scatter(x= df['budget'], y=df['gross'])

plt.title('Budget Vs Gross Earnings')

plt.xlabel('Budget')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#check company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes #random numerization
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#unstacking

correlation_matric = df_numerized.corr()
corr_pairs = correlation_matric.unstack()
corr_pairs.head()
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
high_corr = sorted_pairs [(sorted_pairs)>0.5]
high_corr.head()
```
